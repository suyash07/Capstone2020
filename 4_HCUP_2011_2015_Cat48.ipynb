### 2015 Q4

```python
pathogens10 = ('J09', 'J10','J11','J11','J12','J13','J14','J15','J16','J17','J18')
```

```python
hcup_2015_q4 = pd.read_csv('data_raw\hcup\WA_SIDC_2015_CORE-q4.csv')
```

```python
hcup_2015_q4_1 = hcup_2015_q4.copy()
```

```python
dxn10 = []
for j in range (1,26):
    dx = str('I10_DX') + str(j)
    dxn10.append(dx)
```

```python
columns = ['AMONTH','ATYPE', 'AYEAR','ZIP']
columns.extend(dxn10)
```

```python
hcup_2015_q4_1 = hcup_2015_q4_1[columns]
```

```python
hcup_2015_q4_3 = pd.DataFrame()
for i in range(1,26):
    i10dxn = str('I10_DX') + str(i)
    hcup_2015_q4_1_filtered = hcup_2015_q4_1[i10dxn].str.startswith(pathogens10, na = False)
    hcup_2015_q4_2_filtered = hcup_2015_q4_1[hcup_2015_q4_1_filtered]
    hcup_2015_q4_3 = hcup_2015_q4_3.append(hcup_2015_q4_2_filtered)
hcup_2015_q4_3 = pd.DataFrame.drop_duplicates(hcup_2015_q4_3)
```
